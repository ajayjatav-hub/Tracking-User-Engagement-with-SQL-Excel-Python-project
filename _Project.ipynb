```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/minutes_and_certificates.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score , mean_squared_error,mean_absolute_error
```

```python
X = pd.DataFrame(df['minutes_watched'])
Y = pd.DataFrame(df['certificates_issued'])

x_train , x_test , y_train, y_test = train_test_split(X ,Y , test_size=0.2 , random_state=365)
```

```python
lm = LinearRegression()
lm
```

```python
lm.fit(x_train ,y_train)
```

```python
y_predict = lm.predict(x_test)
lm.coef_
display(r2_score(y_test, y_predict))
mean_absolute_error(y_test, y_predict)
```

The R-squared value for this model is 0.45

**R-squared Interpretation:**

R-squared (R²) is a statistical measure that represents the proportion of the variance in the dependent variable (certificates issued) that is predictable from the independent variable (minutes watched).

In this case, an R-squared of  0.45 means that approximately  45% of the variation in 'certificates\_issued' can be explained by the linear relationship with 'minutes\_watched'. This indicates that 'minutes\_watched' is a predictor of 'certificates\_issued', but it doesn't account for all the variability, suggesting other factors also play a role. An R-squared of 0.45 is generally considered to indicate a moderate fit for a simple linear regression model.

```python

print(f'The intercept is {lm.intercept_} and coefficient is {lm.coef_}')
```

**Model Interpretation:**

Based on the linear regression model, the estimated relationship between `minutes_watched` and `certificates_issued` is:

$certificates\_issued =  1.19452283 \times minutes\_watched + 0.00175484 $

*   **Intercept (1.19452283)**: This is the predicted number of certificates issued when a student has watched 0 minutes. In this context, it represents a baseline number of certificates a student might have, even with minimal or no minutes watched in this specific period.
*   **Coefficient (0.00175484)**: This indicates that for every one-unit increase in `minutes_watched`, the predicted number of `certificates_issued` increases by approximately  0.00175484. This suggests a positive linear
relationship between the time spent watching content and the number of certificates earned.

```python
import matplotlib.pyplot as plt # Import matplotlib for adding labels and legend

plt.figure(figsize=(10, 6)) # Add a figure to make sure the plot is displayed
axl = sns.scatterplot(x=y_test.values.flatten(), y=y_predict.flatten(), color='blue', label='Predicted')
sns.scatterplot(x=y_test.values.flatten(), y=y_test.values.flatten(), color='red', label='Actual', ax = axl)
plt.xlabel('Actual Certificates Issued')
plt.ylabel('Predicted Certificates Issued')
plt.title('Actual vs. Predicted Certificates Issued')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.kdeplot(y_test.values.flatten(), label='Actual', color='b', fill=True, alpha=0.5)
sns.kdeplot(y_predict.flatten(), label='predict', color='r', fill=True, alpha=0.5)
plt.legend()
plt.xlabel('Certificates Issued')
plt.ylabel('Density')
plt.title('Actual vs. Predicted Certificates Issued Distribution')
plt.show()
```
